```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 2.0
sigma = 0.5

norm_rv = sts.norm(loc=mu, scale=sigma)

norm_rv.rvs(size=10)
```

```python
x = np.linspace(0,4,100)
cdf = norm_rv.cdf(x) 
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
binomial_rv = sts.binom(20, 0.7)
binomial_rv.rvs(10)
```

```python
x = np.linspace(0,20,21)
cdf = binomial_rv.cdf(x)
plt.step(x, cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,20,21)
pmf = binomial_rv.pmf(x)
plt.plot(x, pmf, 'o')

plt.ylabel('$P(X=x)$')
plt.xlabel('$x$')
```

```python
poisson_rv = sts.poisson(5)
poisson_rv.rvs(10)
```

```python
x = np.linspace(0,30,31)
for l in [1, 5, 10, 15]:
    rv = sts.poisson(l)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label="$\lambda=%s$" % l)
plt.legend()
plt.title("CDF (poisson)")

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
